```python
import obspy
from obspy.clients.fdsn import Client
import numpy as np

#out = open('Meta_Sensitivity.txt', 'w')
#out.write(f"Net.Station,Channel,Starttime,Endtime,Calc.Sens.,Rep.Sens.,Sens.Differ(%)\n")

net = "IA"
sta = "*"
chan = "HNZ"

client = Client("https://geof.bmkg.go.id", user='infodini', password='$Infodini2022!')
stations = client.get_stations(network=net, 
                          station=sta,
                          channel=chan, 
                          level='station', 
                          location='*')

for i in range(len(stations[0])):
    inv = client.get_stations(network=stations[0].code, 
                          station=stations[0][i].code,
                          channel='*', 
                          level='response', 
                          location='*')

    
    net = inv[0]
    sta = inv[0][0]
    print(f"{net.code}.{sta.code}")
    for i in range(len(sta)):
        #print(f"Working on Channel: {inv[0][0][i].code}, Time: {inv[0][0][i].start_date}-{inv[0][0][i].end_date}")
        try:
            cha = inv[0][0][i]
            start = cha.start_date
            end = cha.end_date
            
            datetime = obspy.UTCDateTime("2025-01-01T00:00:00")
            response = cha.response
            f = float(response.instrument_sensitivity.frequency)
            
            #calc_sens = response.get_evalresp_response_for_frequencies([f], 
            #                                                  output='VEL', 
            #                                                  hide_sensitivity_mismatch_warning=False)
            rep_sens = response.instrument_sensitivity.value
            response.recalculate_overall_sensitivity(frequency=f)
            calc_sens = response.instrument_sensitivity.value

            if calc_sens and rep_sens and end == None :
                delta = abs(calc_sens - rep_sens)
                delta_percent = (delta/calc_sens) * 100
                #print(f"{net.code}.{sta.code}.{cha.location_code}.{cha.code}")
                if delta_percent >= 5.0 :
                    print(f"{net.code}.{sta.code}.{cha.location_code}.{cha.code}")
                    print(f"Calculated Sensitivity: {calc_sens} Counts/m/s at {f} Hz")
                    print(f"Reported Sensitivity: {rep_sens} Counts/m/s at {f} Hz")
                #    if end == None :
                    print(f"Calculated and Reported sensitivity differ: {delta_percent}")
                #        out.write(f"{net.code}.{sta.code},{cha.code},{start},{end},{abs(calc_sens)[0]},{rep_sens},{delta_percent}\n")
                #else:
                    #out.write(f"{net.code}.{sta.code}  {cha.code}  {start}  {end}  {abs(calc_sens)[0]}  {rep_sens}  {delta_percent}\n")
                    
        except Exception as e :
            #if end == None :
            print(f"Working on Channel: {cha.code}, Time: {start}-{end}")
                #out.write(f"{net.code}.{sta.code},{cha.code},{start},{end},{e}\n")
            print(e)
            continue
    print(' ')
    print('################################################################################')
    print(' ')

#out.close()
```
